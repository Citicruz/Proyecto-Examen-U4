# **MARCO TERICO**

***Prueba ANOVA***

La prueba ANOVA (Anlisis de la Varianza) es una tcnica estadstica utilizada para comparar las medias de tres o ms grupos independientes y determinar si al menos uno de esos grupos es significativamente diferente de los dems. A travs de este anlisis, se evala si las diferencias observadas en las medias de los grupos son mayores de lo que se esperara por casualidad.
ANOVA genera un valor F (estadstico F), que es la razn entre la variabilidad entre los grupos y la variabilidad dentro de los grupos. Si el valor de F es mayor que un valor crtico determinado por una tabla de distribucin F, se rechaza la hiptesis nula, lo que sugiere que existen diferencias significativas entre las medias de los grupos.

***Prueba Tukey***

La prueba de Tukey, tambin conocida como prueba de Tukey HSD (Honest Significant Difference), es una tcnica estadstica utilizada para realizar comparaciones por pares entre las medias de varios grupos despus de haber obtenido un resultado significativo en una prueba ANOVA.
Su principal objetivo es identificar cules grupos son significativamente diferentes entre s, controlando el error tipo I (la probabilidad de rechazar incorrectamente una hiptesis nula verdadera) cuando se realizan mltiples comparaciones. Cuando ANOVA muestra que hay diferencias significativas entre los grupos, la prueba de *Tukey* se aplica para examinar qu grupos especficamente difieren.

*Valor p*

El valor p es una herramienta fundamental en la estadstica inferencial que ayuda a determinar si los resultados de un estudio son lo suficientemente fuertes como para rechazar la hiptesis nula. Su interpretacin depende de la comparacin con el umbral de significancia, y su valor refleja la probabilidad de obtener los resultados observados bajo la suposicin de que la hiptesis nula es cierta.

```python
import pandas as pd
import scipy.stats as stats

df = pd.read_csv('https://raw.githubusercontent.com/Citicruz/Proyecto-Examen-U4/refs/heads/main/dataset.csv')
df.drop(columns=[], inplace=True)
df.dropna(inplace=True)
df
```

```python
df.columns
```

```python
df['participacion_parental'].unique()
```

```python
# Si todos los grupos tienen una dristribucin normal
import scipy.stats as stats

nivel_de_significancia = 0.05

# Realizar ANOVA
anova_result = stats.f_oneway(
    df[df['participacion_parental'] == 'bajo']['puntaje_examen'],
    df[df['participacion_parental'] == 'medio']['puntaje_examen'],
    df[df['participacion_parental'] == 'elevado']['puntaje_examen'],
)

# Interpretar los resultados
print("Resultados del ANOVA:")
print(f"F-Estadstico: {anova_result.statistic:.2f}")
print(f"Valor p: {anova_result.pvalue:}\n")

if anova_result.pvalue < nivel_de_significancia:
    print("Conclusin: Hay diferencias significativas entre al menos dos tratamientos.")
else:
    print("Conclusin: No hay diferencias significativas entre los tratamientos.")
```

```python
from statsmodels.stats.multicomp import pairwise_tukeyhsd
import matplotlib.pyplot as plt

nivel_de_significancia = 0.05

# Prueba de Tukey
tukey = pairwise_tukeyhsd(endog=df['puntaje_examen'], groups=df['participacion_parental'], alpha=nivel_de_significancia)

# Mostrar los resultados
print(tukey)

# Grfico de las diferencias entre grupos
tukey.plot_simultaneous(ylabel="Grupos", xlabel="Medias")

plt.gca().spines['right'].set_visible(False) # derecha
plt.gca().spines['top'].set_visible(False)   # superior
plt.title("Mltiples comparaciones entre todos los pares (Tukey)")
```

**CONCLUSIN**

Con base con los resultados de la prueba ANOVA se puede observar que existen diferncias significativas entre los grupos de los datos debido a que el valor p de la prueba sobre pasa el asignado de 0.05 dandonos un resultado de 3.08 lo que nos lleva a la prueba Tukey la cual nos da como resultado que los tres pares de esta parte si cumplen con diferencias significativas.
